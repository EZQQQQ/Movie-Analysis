DATA CLEANING

```python
# Basic Libraries
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt 
%matplotlib inline
import json 
```

IMPORT DATA SET

```python

raw_data = pd.read_csv('movie1.csv')
raw_data.head(50)
```

```python
# Inforamtion about the Variables
raw_data.info()
```

```python
# describe our data
raw_data.describe()
```

Drop unused columns in the analysis


```python
raw_data.drop(['ID', 'FILMID', 'TITLE', 
               'ADULT', 'BACKDROP_PATH','HOMEPAGE','IMDB_ID','ORIGINAL_TITLE', 'OVERVIEW',
               'POSTER_PATH','STATUS_','TAGLINE','COLLECTIONID','ORIGINAL_LANGUAGE'],axis=1,inplace=True)
raw_data.head(20)
```

Drop zero value row

```python
print("Rows with none zero values in Budget: ", raw_data[(raw_data['BUDGET']!=0)].shape[0])
print("Rows with none zero values in Revenue: ", raw_data[(raw_data['REVENUE']!=0)].shape[0])
print("Rows with none zero values in Vote_count: ", raw_data[(raw_data['VOTE_COUNT']!=0)].shape[0])
print("Rows with none zero values in POPULARITY: ", raw_data[(raw_data['POPULARITY']!=0)].shape[0])


# DROP data with value less than 0
raw_data = raw_data.loc[(raw_data[['REVENUE',"BUDGET", "VOTE_COUNT", "POPULARITY"]]>0).all(axis=1)]
```

```python
# describe our data
raw_data.head()
```

SPILT RELEASE DATE INTO YEAR AND MONTH

```python
raw_data['RELEASE_DATE'] = pd.to_datetime(raw_data['RELEASE_DATE'])
raw_data['RELEASE_DATE'].head(50)
```

```python
# Extract year and month of release date and Put into new columns
raw_data['release_year'] = raw_data['RELEASE_DATE'].dt.year
raw_data['release_month'] = raw_data['RELEASE_DATE'].dt.month

# Drop release_date columns
raw_data.drop(['RELEASE_DATE'],axis=1,inplace=True)
raw_data = raw_data.reset_index(drop=True)

# Fill NA/NaN values using 0
raw_data['release_year'] = raw_data['release_year'].fillna(0)
raw_data['release_month'] = raw_data['release_month'].fillna(0)

# Convert dtypes of columns
raw_data['release_year'] = raw_data['release_year'].astype('int64')
raw_data['release_month'] = raw_data['release_month'].astype('category')

display(raw_data['release_year'].describe())
display(raw_data['release_month'].describe())
```

```python
raw_data.head(100)
```

Drop movies that are produced earlier than 1990

```python
print("Rows with year under 1990: ", raw_data[(raw_data['release_year']<1990)].shape[0])


# DROP data with value less than 0
raw_data = raw_data.loc[raw_data['release_year']>=1990]
raw_data = raw_data.reset_index(drop=True)
```

```python
raw_data.describe()
```

```python
#inflation rate from 1990 to 2022
#Inflation rate from https://www.calculator.net/inflation-calculator.html?cstartingamount1=100&cinyear1=1990&coutyear1=2022&calctype=1&x=93&y=21
inflation = [2.1876, 2.0757, 1.9911, 1.9326 ,1.877, 1.8292, 1.7793, 1.7286, 1.6891, 1.6633,  1.6277,
             1.5744, 1.5311, 1.5071, 1.4737, 1.4352, 1.3882, 1.3446, 1.3073, 1.2589, 1.2632, 1.2428,
             1.2047, 1.1803, 1.1632, 1.1456, 1.1433, 1.129, 1.1055, 1.0792, 1.06, 1.047, 1]
```

#Adjust revenue according to inflation

```python
adjusted_revenue = []
adjusted_budget = []
for index, rows in raw_data.iterrows():
    adjusted_budget.append(int(rows['BUDGET'] * inflation[round(rows['release_year']-1990)]))
    adjusted_revenue.append(int(rows['REVENUE'] * inflation[round(rows['release_year']-1990)]))
    
    
adjusted_budget = pd.DataFrame(adjusted_budget, columns = ['New_Budget'])
adjusted_revenue = pd.DataFrame(adjusted_revenue, columns = ['New_Revenue'])

raw_data = pd.concat([raw_data, adjusted_budget, adjusted_revenue], axis = 1)
```

```python
raw_data.head()
```

#Round up/down the vote_average to whole number

```python
adjusted_vote_avg=[]
for index, rows in raw_data.iterrows():
    adjusted_vote_avg.append(round(rows['VOTE_AVERAGE']))

adjusted_vote_avg= pd.DataFrame(adjusted_vote_avg, columns=["New_Vote_Avg"])

raw_data = pd.concat([raw_data, adjusted_vote_avg], axis = 1)
raw_data.head(100)
```

EXPORT DATA

```python
raw_data.to_csv('clean_data.csv', header=True, index=False)
```
